```python
white_wine['type'] = 'white'
red_wine['type'] = 'red'
```

```python
wines_df = pd.concat([red_wine,white_wine])
```

```python
sns.countplot(x = 'quality',hue = 'type' , data = wines_df)
```

```python
sns.displot(data = red_wine, x = 'quality',kind = 'kde',color = 'r', label = 'red wine')
```

```python
p1 = sns.kdeplot(red_wine['quality'],shade = True,color = 'r', label =  'red wine')
p2 = sns.kdeplot(white_wine['quality'],shade = True,color = 'b' , label = 'white wine')
```

```python
sns.boxplot(data = wines_df , x = 'quality' , y = 'alcohol' , hue = 'type' , palette = ["r","w"] ,showfliers = False )
```

```python
sns.boxplot(data = wines_df , x = 'quality' , y = 'density' , hue = 'type' , palette = ["r","w"], showfliers = False )
```

```python
sns.boxplot(data = wines_df , x = 'quality' , y = 'residual sugar' , hue = 'type' , palette = ["r","w"], showfliers = False )
```

```python
s1 = sns.kdeplot(red_wine['residual sugar'] , shade = True , color = 'red', label = 'red wine' )
s2 = sns.kdeplot(white_wine["residual sugar"],shade  =True, color = 'green', label = 'white wine')
```
