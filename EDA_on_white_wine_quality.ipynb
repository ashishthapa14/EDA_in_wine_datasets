```python
from scipy.stats import skew

def plot_hist(temp_df , bin_size = 25):
    ax = sns.distplot(temp_df)
    #xmin ,xmax = ax.get_xlim()
    #ax.set_xticks(np.round(np.linspace(xmin, xmax, bin_size), 2))
    plt.tight_layout()
    plt.locator_params(axis = 'y', nbins = 6)
    plt.show()
    ## central tendency
    print(f"Mean is {temp_df.mean()}")
    print(f"Median is {temp_df.median()}")
    print(f"Mode is {temp_df.mode()}")
    
    #skewness :-
    print(f"skewness is {temp_df.skew()}")
```

```python
plot_hist(white_wine['alcohol'])
```

```python
def plot_box(X,Y,temp_data):
    ax = sns.boxplot(x = X,y=Y,data = temp_data,showfliers = False)
    #xmin ,xmax = ax.get_xlim()
    #ax.set_xticks(np.round(np.linspace(xmin, xmax, bin_size), 2))
    plt.show()
```
